```python
import pandas as pd
import numpy as np
```

```python
import timeit

start = timeit.default_timer()

data = pd.read_csv("data/lowz_test_sdss.csv")
```

```python
df = pd.DataFrame(title)
df=df.transpose()
print(df)
```
